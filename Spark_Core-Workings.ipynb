```python
#Q2. Consider the dataset file Olympics.csv. This file contains information about the Olympic games, players participating in the games, and details of medals won by them. Using Spark core and the data file, compute the following:

#a)	Total medals that each country won in a particular sport (such as Gymnastics).
#b)	In each Olympic games, how many medals has India won? 
#c)	Compute top 3 countries in terms of total medals by each Olympic games year.
```

# Q2

```python
olympicsrdd = sc.textFile("./olympics.csv")
```

```python
#olympicsrdd.take(1)
```

```python
olympicsrdd_split = olympicsrdd.map(lambda line: line.split('\t'))
```

```python
#olympicsrdd_split.take(1)
```

## Q2 (a)

```python
#olympicsrdd_split_kv = olympicsrdd_split.map(lambda z: (z[2],z[5],(z[0],z[1],z[3],z[4],z[6],z[7],z[8],z[9])))
olympicsrdd_split_kv = olympicsrdd_split.map(lambda z: [(z[2],z[5]),(int)(z[9])])
#olympicsrddCtrywise = olympicsrdd_split.groupBy(lambda w: w[2,5])
#olympicsrdd_split_kv.collect()
```

```python
print("Q2(a) Total medals that each country won in any sport is: ")
```

```python
#olympicsrddGrp = olympicsrdd_split.groupBy(lambda k: (k[2],k[5]) )
olympicsrdd_split_kv.reduceByKey(lambda x,y: x+y).collect()
#olympicsrddGrp.collect()
#print ([(k, list(v)) for (k, v) in olympicsrddGrp.collect()])
```

## Q2 (b)

```python
olympicsrdd_India = olympicsrdd_split.filter(lambda x: x[2] == "India").map(lambda x: (x[3],(int)(x[9])))
#olympicsrdd_India.collect()
```

```python
print("Q2(b) Count of medals India won in each Olympics game is: ")
```

```python
olympicsrdd_India.reduceByKey(lambda x,y:x+y).collect()
```

## Q2 c)

```python
olympicsrddCtryYearGrp = olympicsrdd_split.map(lambda z: [(z[2],z[3]),int(z[9])])
#olympicsrdd_split.groupBy(lambda k: (k[2],k[3]) )
#olympicsrddCtryYearGrp.collect()
#print ([(k, list(v)) for (k, v) in olympicsrddCtryYearGrp.collect()])
```

```python
#olympicsrddCtryYearGrp.reduceByKey(lambda x,y:x+y).sortBy(lambda x: x[1], ascending = False).collect()
olympicsrddCtryYearSrt = olympicsrddCtryYearGrp.reduceByKey(lambda x,y:x+y)#.sortBy(lambda x: x[1], ascending = False)
#olympicsrddCtryYearSrt.take(3)
```

```python
print("Q2(c) Top 3 countries in terms of total medals by each Olympic games year is: ")
```

```python
olympicsrddCtryRemap = olympicsrddCtryYearSrt.map(lambda x: ((x[0][1]),(x[1],x[0][0])))
#olympicsrddCtryRemap.collect()
olympicsrddCtryRemap.groupByKey().map(lambda y:((y[0]),sorted(y[1],reverse=True)[:3])).collect()
```
